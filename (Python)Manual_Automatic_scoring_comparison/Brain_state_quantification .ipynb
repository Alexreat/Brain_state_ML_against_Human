```python
dfs = [animal_129,animal_132,animal_139,animal_227,animal_229,animal_236,animal_237,animal_238,animal_241,animal_365,animal_366,animal_369,animal_373,animal_382,animal_404,animal_433]
counts_dict = {}
#iterate over the dataframes and add up the values couts 
for df in dfs:
    for index, value in df['brainstate'].value_counts().items():
        counts_dict[index] = counts_dict.get(index,0) + value

#conversion of the dictionary to a pandas series 
counts = pd.Series(counts_dict)

print(counts)
```

```python
# create a sample series with the int64index as index
s = pd.Series([19525, 10858, 11692, 976, 2972], index=pd.Index([1, 5, 0, 2, 4], dtype='int64'))

# create a dictionary to map the old index labels to new labels
index_map = {0: 'Wake', 1: 'Nonrem', 2: 'Rem', 4: 'Seizure', 5: 'Noise'}

# rename the index labels using the map
s = s.rename(index=index_map)

# print the updated series
print(s)
```

```python
# drop the 'Noise' variable from the series
s = s.drop('Noise')
s = s.drop('Seizure')

# set the size of the figure
plt.figure(figsize=(10, 6))

# create a bar plot of the series
ax = s.plot.bar()

# set the title and axis labels
ax.set_title('Distribution of brain states')
ax.set_xlabel('Brain states')
ax.set_ylabel('Count')

# adjust the layout
plt.tight_layout()

# save the plot to a file
plt.savefig('brain_states.png')

# show the plot
plt.show()
```
